```python
import numpy as np
import matplotlib.pyplot as plt
from pyanom.outlier_detection import HotelingT2
```

## Prepare data

```python
X_normal = np.loadtxt("./data/normal_data.csv", delimiter=",")
X_error = np.loadtxt("./data/error_data.csv", delimiter=",")

print(X_normal.shape)
print(X_error.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3), sharey=True)

ax[0].plot(X_normal)
ax[1].plot(X_error)
plt.show()
```

## Training

```python
model = HotelingT2()
model.fit(X_normal)
```

## Inference

```python
anomaly_score = model.score(X_error)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 6), sharex=True)
ax[0].plot(X_error)
ax[1].plot(anomaly_score)
plt.show()
```
