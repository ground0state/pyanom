```python
import numpy as np
import matplotlib.pyplot as plt
from pyanom.outlier_detection import CAD
```

## Prepare data

```python
y_normal = np.loadtxt(
    "./data/timeseries_normal.csv", delimiter=",").reshape(-1, 1)
y_error = np.loadtxt(
    "./data/timeseries_error.csv", delimiter=",").reshape(-1, 1)

print(y_normal.shape)
print(y_error.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3), sharey=True)

ax[0].plot(y_normal)
ax[1].plot(y_error)
plt.show()
```

## Training

```python
model = CAD(threshold=1.0)
model.fit(y_normal)
```

## Inference

```python
anomaly_score = model.score(y_error)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 6), sharex=True)
ax[0].plot(y_error)
ax[1].plot(anomaly_score)
plt.show()
```
