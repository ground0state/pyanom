```python
import numpy as np
import matplotlib.pyplot as plt
from pyanom.subspace_methods import SSA
```

## Prepare data

```python
y_error = np.loadtxt("./data/timeseries_error2.csv", delimiter=",")

print(y_error.shape)
```

```python
fig, ax = plt.subplots(1, figsize=(5, 3))

ax.plot(y_error)
plt.show()
```

## Training

```python
model = SSA(window_size=50, trajectory_n=25, trajectory_pattern=3, test_n=25, test_pattern=2, lag=25)
model.fit(y_error)
```

## Inference

```python
anomaly_score = model.score()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 6), sharex=True)
ax[0].plot(y_error)
ax[1].plot(anomaly_score)
plt.show()
```
